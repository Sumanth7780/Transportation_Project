```python
import pandas as pd

# Load the dataset
file_path = "D:\\0-transportation project\\code\\for_100\\common_pcards\\common_pcards_00.csv"

data = pd.read_csv(file_path)

# Display the first few rows of the dataframe to understand its structure
data.head()
```

```python
# Randomly sample 100 users from the dataset
sampled_users = data

# Display the sampled users
sampled_users
```

```python
sampled_users.to_csv("D:\\0-transportation project\\code\\for_100\\sample_100.csv", index=False)
```

```python
def filter_data_based_on_pcards(sampled_users, data_df, output_file_path=None):
    """
    Filters data based on a list of common PCards.

    Parameters:
    - common_pcards_path: str, path to the CSV file containing common PCards.
    - data_file_path: str, path to the CSV file containing data to be filtered.
    - output_file_path: str, optional, path to save the filtered data as a CSV file.

    Returns:
    - DataFrame containing filtered data.
    """


    # Filter the data
    filtered_data = data_df[data_df['PCARD_NO'].isin(sampled_users)]

    # Save the filtered data to a CSV file if an output path is provided
    if output_file_path:
        filtered_data.to_csv(output_file_path, index=False)
        print(f"Filtered data saved to {output_file_path}")

    return filtered_data
```
